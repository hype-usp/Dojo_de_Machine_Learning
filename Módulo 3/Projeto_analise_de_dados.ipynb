#Projeto de Análise de Dados

Em duplas, enfrentem o desafio deste projeto de realizar uma Análise Exploratória em um conjunto de dados. A análise exploratória é a parte do projeto de Aprendizado de Máquina que nos permite compreender o comportamento geral dos dados. Com essa compreensão, podemos planejar nosso modelo melhor e utilizar as variáveis mais explicativas, com o objetivo de maximizar a taxa de acerto.

Para que um modelo de Aprendizado de Máquina obtenha bons resultados, é necessário que os dados de entrada tenham qualidade. Ou seja, devem ser dados que expliquem a variável que estamos tentando prever.

As ferramentas de manipulação e visualização de dados vistas nos últimos módulos podem ser utilizadas para identificar quais variáveis possuem mais poder explicativo e como se dá a sua relação com o alvo a ser previsto.

Neste módulo, vocês vão aplicar os conceitos aprendidos nos últimos dois.

Trouxemos 5 sugestões de datasets:


- [Iris](https://www.kaggle.com/datasets/uciml/iris) (Classficação de especies de plantas)
-[Titanic](https://www.kaggle.com/competitions/titanic/data?select=train.csv) (Lista de pessoas que estavam no Titanic)
- [Stellar Classification](https://www.kaggle.com/datasets/fedesoriano/stellar-classification-dataset-sdss17) (Classificação de tipos de estrelas)
-[Heart Failure Prediction](https://www.kaggle.com/datasets/andrewmvd/heart-failure-clinical-data) (Predição de problema cardíaco)
-[Data Science Salaries](https://www.kaggle.com/datasets/arnabchaki/data-science-salaries-2023) (Salários de profisionais de datascience)

Estes datasets foram escolhidos por serem muito famosos no meio de datascience e sua relativa facilidade de serem trabalhados, mas vocês podem escolher qualquer outro conjunto de dados que tiverem interesse de trabalhar. Só pedimos que, se escolherem outro dataset, avisem algum dos monitores no Telegram e enviem o seu Link para que possamos garantir que o dataset servirá para o propósito desse projeto.

Seu objetivo aqui é fazer uma análise exploratória em algum desses datasets. Utilizem plotagens, operações, métodos de síntese de dados, e qualquer outra ferramenta que lhes permita extrair insights do conjunto de dados.


Note que todos estes conjuntos de dados possuem uma coluna "Alvo", que seria a coluna a ser prevista caso estivessemos fazendo um projeto de machine learning. Entenda o conjunto de dados como um todo, mas direcione a sua análise para a coluna alvo. É a coluna mais importante do conjunto de dados em um projeto de Machine Learning

Elementos esperados no projeto:
- importação correta do dataset
- seleção de linhas e colunas
- filtro de linhas
- plotagens de variáveis únicas
- plotagens de pares de variáveis
- medidas descritivas das variáveis
- além do notebook, a dupla deve preparar uma apresentação em vídeo em que ambas as pessoas expliquem por que escolheram este data set e o processo de análise de dados. Não se limitem a plotar resultados, mas tentem extrair alguma interpretação destes de acordo com o fenômeno estudado. Por exemplo: pessoas que pagaram mais para entrar no Titanic tiveram mais chances de sobrevivência? Existe relação demonstrável entre tamanho da empresa e salário dos profissionais de data science?

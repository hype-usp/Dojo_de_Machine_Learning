#Projeto De ML

Estamos na última fase do Dojo. Nos últimos módulos vocês viram ferramentas de extração de informaçãoes de dados e tiveram uma introdução conceitual ao Machine Learning.

Agora, seu objetivo será aplicar na prática os conhecimentos adiquiridos e montar um modelo de Machine Learning que seja capaz de predizer alguma coluna específica de um conjunto de dados.

Cada dupla está livre para escolher qualquer dataset que quiser. Mas, sugerimos que utilize um dos 5 que foram passados no case de análise de dados. Pode ser o mesmo já utilizado na primeira entrega.

Link dos datasets sugeridos:

- [Iris](https://www.kaggle.com/datasets/uciml/iris) (Classficação de especies de plantas)
-[Titanic](https://www.kaggle.com/competitions/titanic/data?select=train.csv) (Lista de pessoas que estavam no Titanic)
- [Stellar Classification](https://www.kaggle.com/datasets/fedesoriano/stellar-classification-dataset-sdss17) (Classificação de tipos de estrelas)
-[Heart Failure Prediction](https://www.kaggle.com/datasets/andrewmvd/heart-failure-clinical-data) (Predição de problema cardíaco)
-[Data Science Salaries](https://www.kaggle.com/datasets/arnabchaki/data-science-salaries-2023) (Salários de profisionais de datascience)
